# Question 1
## what we are going to do :
build a matrix A and vector b such that Ax = b, where the data is given in the assignment page. 
We use numpy to build the matrix and vector, and we will try to solve the following problems.
The first problem is just building this linear system.

```python
# Question 1-(a)
import numpy as np
A = np.array([[2,1,0,-4],[1,0,0.25,-1],[-2,-1.1,0.25,6.2],[4,2.2,0.3,-2.4]])
b = np.array([-3,-1.5,5.6,2.2])
print("Matrix A:")
print(A)
print("Vector b:")
print(b)
```

We can see the matrix and vector above.

Then, in problem 1-b, we want to solve the system with numpy.linalg.

```python
# Question 1-(b)
x = np.linalg.solve(A, b)
x
```

We can see the answer above. x1=0.5, x2=2, x3=-2, x4=1.5.

Then, in problem 1-c, we want to find the inverse of matrix A, i.e., A^(-1), 
with numpy.linalg. We will calculate A^(-1)b, too.

```python
# Question 1-(c)
A_inv = np.linalg.inv(A)
print("Inverse of matrix A:")
print(A_inv)

x = np.dot(A_inv, b)
print("Solution:")
print(x)
```

the matrix A^(-1) is shown above.

We can see the answer above. x1=0.5, x2=2, x3=-2, x4=1.5, which is the same as the Ax=b. We can see the answer of A^(-1)x=b is identical as Ax=b.

When we solve the linear system Ax = b using the numpy.linalg.solve function, we are directly finding the values of x that satisfy the equation Ax = b. On the other hand, when we find A^(-1)*b, we are first finding the inverse of the matrix A and then multiplying it with the vector b to obtain the solution vector x.
However, mathematically, we know that if we multiply both sides of the equation Ax = b by A^(-1), we get:

A^(-1)Ax = A^(-1)*b

Simplifying the left-hand side, we get:

Ix = A^(-1)*b

where I is the identity matrix. Therefore, we get:

x = A^(-1)*b

This shows that the solution obtained by finding A^(-1)*b is the same as the solution obtained by solving the linear system Ax = b

Finally, we want to find the LU-factorization of the coefficient matrix of system Ax=b:

```python
# Question 1-(d)

import scipy
p, l, u = scipy.linalg.lu(A)
# Print the factors
print("Permutation matrix P:")
print(p)
print("Lower-triangular matrix L:")
print(l)
print("Upper-triangular matrix U:")
print(u)
```

We can see the Permutation matrix P, Lower-triangular matrix L, and Upper-triangular matrix U.
To check whether A = PLU, we try to dot them together.

```python
print("Product of PLU:")
print(np.dot(p, np.dot(l, u)))
```

We can see that the answer is composed correctly.
